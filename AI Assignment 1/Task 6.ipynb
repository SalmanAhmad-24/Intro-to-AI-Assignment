```python
from queue import Queue

def is_valid(board, row, col, n):
    # Check if there is a queen in the same row
    for i in range(col):
        if board[row][i] == 1:
            return False
    
    # Check if there is a queen in the upper left diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    
    # Check if there is a queen in the lower left diagonal
    i, j = row, col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1
    
    # If no conflicts, then the placement is valid
    return True

def bfs_n_queen(n):
    queue = Queue()
    queue.put([])
    
    while not queue.empty():
        board = queue.get()
        
        # If the board is complete, return it
        if len(board) == n:
            return board
        
        # Otherwise, generate all possible next moves and add them to the queue
        for i in range(n):
            if is_valid(board, i, len(board), n):
                new_board = board.copy()
                new_board.append(i)
                queue.put(new_board)
    
    # If no solution found, return empty list
    return []

# Example usage:
n = 8
solution = bfs_n_queen(n)

if solution:
    print("Solution found:")
    for row in range(n):
        line = ""
        for col in range(n):
            if solution[row] == col:
                line += "Q "
            else:
                line += ". "
        print(line)
else:
    print("No solution found.")
```

```python
from collections import deque

def valid(state, new_queen):
    for col, row in enumerate(state):
          if abs(row - new_queen) == abs(col - len(state)) or\
          new_queen in state:
              return False        
    return True

size = 8
queue = deque((queen_pos,) for queen_pos in range(size))
while queue:
    board_state = queue.popleft()
    if len(board_state) == size:
        print('solution:', board_state)
        continue
    for queen_pos in range(size):
        if not valid(board_state, queen_pos):
            continue
        next_state = board_state + (queen_pos,)
        queue.append(next_state)
```
